# Task 1 : Fetching Data

```python
import requests
s=requests.get("https://www.python.org/")
s.text
```

```python
import bs4
j=bs4.BeautifulSoup(s.text,"lxml")
print(j)
```

```python
b=j.getText()
```

```python
# open in text file
f=open("web_python.txt","w",encoding="utf-8")
f.write(b)
f.close()
```

# Task 2 : Finding Web Frequency

```python
# Read in text file
f=open("web_python.txt","r")
f.read()
f.close()
```

```python
# web frequency
freq_dict={}
count=1
s=b.split()

for i in range (len(s)):
    if s[i].isalpha():
        if s[i] in freq_dict:
            count+=1
            freq_dict[s[i]] =(count)
        
        else:
            freq_dict[s[i]] =(count)
# final output
f=open("b.txt","w")
f.write(str(freq_dict))
f.close()


```
